```python
# Cellule 1 : Connexion à la base DWH locale
import pandas as pd
import sqlalchemy
import os

db_path = os.path.abspath('../data/warehouse/northwind_dwh.db')
engine = sqlalchemy.create_engine(f'sqlite:///{db_path}')

print(f"Connexion établie à : {db_path}")

# Cellule 2 : Comparaison Row Count (Raw vs DWH)
# On compare le nombre de lignes dans Details (Raw) vs FactSales (DWH)
raw_details = pd.read_csv('../data/raw/Order_Details.csv') # ou 'Order Details.csv'
dwh_fact = pd.read_sql("SELECT * FROM FactSales", engine)

count_raw = len(raw_details)
count_dwh = len(dwh_fact)

print(f"Lignes Raw (Details) : {count_raw}")
print(f"Lignes DWH (Fact)    : {count_dwh}")

if count_raw == count_dwh:
    print("✅ TEST SUCCESS : Le nombre de lignes correspond.")
else:
    print(f"⚠️ ATTENTION : Différence de {abs(count_raw - count_dwh)} lignes.")

# Cellule 3 : Vérification des doublons
duplicates = dwh_fact.duplicated().sum()
print(f"Doublons dans FactSales : {duplicates}")

# Cellule 4 : Vérification des valeurs Nulles critiques
nulls = dwh_fact[['DateKey', 'CustomerKey', 'TotalAmount']].isnull().sum()
print("\nValeurs nulles dans les clés étrangères :")
print(nulls)
```
